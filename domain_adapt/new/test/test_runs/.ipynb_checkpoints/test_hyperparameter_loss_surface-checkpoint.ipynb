```python
# define fitter
options = {'maxiter':5}
def get_fitter(c_B, c_logreg, tradeoff, sigma):
    return constructors.logreg_ratio_1d_unconstrained_fitter(c_B, c_logreg, tradeoff, sigma, options=options)
```

```python
# define data
x_dim = 50
num_train = 20
num_test = 20
proportion = 0.5
def get_data(seed):
    np.random.seed(seed)
    return data.why_weighting_is_important_data(x_dim=x_dim, num_train=num_train, num_test=num_test, proportion=proportion)
```

```python
# define hyperparameter search space via an iterator
c_Bs = [0.1,1.]
c_logregs = [0.1]#,1.]
tradeoffs = [0.001]#,0.01,0.1]
sigmas = [0.25]#,1.,5.]
hyperparams = itertools.product(c_Bs, c_logregs, tradeoffs, sigmas)
```

```python
# define horse that takes in hyperparameter, outputs dictionary of results
def horse_helper(get_data, get_fitter, num_iterations, hyperparams):
    fitter = get_fitter(*hyperparams)
    test_losses = []
    for i in xrange(num_iterations):
        xs_train, xs_test, ys_train, ys_test = get_data(seed=i)
        predictor = fitter(xs_train, xs_test, ys_train)
        ys_test_hat = predictor(xs_test)
        test_errors = ys_test - ys_test_hat
        test_loss = np.dot(test_errors, test_errors)
        test_losses.append(test_loss)
    return {'mean_loss':np.mean(test_losses),'std_loss':np.std(test_losses),'test_losses':test_losses}

num_iterations = 1
horse = functools.partial(horse_helper, get_data, get_fitter, num_iterations)
```
